```python
import random
import tkinter as tk
from tkinter import messagebox

global user_score, computer_score
user_score = 0
computer_score = 0

def ask_continuation():
    continuation_reply = messagebox.askyesno(title = "Play again?", message = "Do you want to play again?")
    if continuation_reply:
        user_input.set("")
        play()
    else:
        root.destroy()

def get_user_choice():
    user_choice = user_input.get()
    if user_choice:
        return user_choice
    else:
        messagebox.showerror("Important", "You did not enter any choice. Please choose one of the three options.")
        return None
     
    

def get_computer_choice():
    game = ["Rock", "Paper", "Scissors"]
    computer_choice = random.choice(game)
    return computer_choice

def winner(user_choice, computer_choice):
    if user_choice == "Rock":
        if computer_choice == "Rock":
            result = "It is a tie!"
        elif computer_choice == "Paper":
            result = "You lose! :("
        else:
            result = "You win! :)"
    elif user_choice == "Paper":
        if computer_choice == "Rock":
            result = "You win! :)"
        elif computer_choice == "Paper":
            result = "It is a tie!"
        else:
            result = "You lose! :("
    else:
        if computer_choice == "Rock":
            result = "You lose! :("
        elif computer_choice == "Paper":
            result = "You win! :)"
        else:
            result = "It is a tie!"
    return result

def score_management(result):
    global user_score, computer_score
    if result == "You lose! :(":
        computer_score += 1
    elif result == "You win! :)":
        user_score += 1
    score_display.config(text = f"Your score = {user_score}\n Computer's score = {computer_score}", font = ("Arial", 8))
    

def play():
    global user_score, computer_score
    user_choice = get_user_choice()
    if user_choice is None:
        return
    computer_choice = get_computer_choice()
    display_choice.config(text = f"Your choice is {user_choice}.\nThe computer's choice is {computer_choice}.", font = ("Arial", 8))
    
    result = winner(user_choice, computer_choice)
    display_result.config(text = f"{result}", font = ("Arial", 8))
    score_management(result)


    
    
    
    
    
    
root = tk.Tk()
root.geometry("900x900")
root.title("Rock-Paper-Scissors Game")
instruction = tk.Label(root, text = "WELCOME!!! \nThis is a Rock-Paper-Scissors Game against the computer. \nTo start the game, choose one of the three options and press the 'Play' button. \nThe interface will save both yours and the computer's score. \nPress 'Continue' at the end of each round. \nThe game will continue until you say 'No' to the pop-up question that asks you whether you want to play another round.\nENJOY!!!", font = ("Arial", 10))
instruction.pack(padx = 10, pady = 10)
user_input = tk.StringVar()
game = ["Rock", "Paper", "Scissors"]
for option in game:
    choice = tk.Radiobutton(root, text = option, variable = user_input, value = option, width = 10, padx = 10, pady = 10)
    choice.pack()
play_btn = tk.Button(root, text = "Play", command = play)
play_btn.pack(padx = 10, pady = 10)

score_display = tk.Label(root, text = f"Your score = {user_score}\n Computer's score = {computer_score}", font = ("Arial", 8))
score_display.pack(padx = 10, pady = 10)

display_choice = tk.Label(root, font = ("Arial", 8))
display_choice.pack(padx = 10, pady = 10)

display_result = tk.Label(root, font = ("Arial", 8))
display_result.pack(padx = 10, pady = 10)

continue_btn = tk.Button(root, text = "Continue", command = ask_continuation)
continue_btn.pack(padx = 10, pady = 10)









root.mainloop()
    
    
```
